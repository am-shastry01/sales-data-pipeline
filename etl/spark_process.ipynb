```python
df = spark.read.csv("D:\\sales-data-pipeline\\data\\sales.csv", header=True, inferSchema=True)

print("Dataset Loaded")

df.show(5)
```

```python
# total revenue by product
product_revenue = df.groupBy("product").sum("revenue")
print("Revenue by Product")
product_revenue.show()
```

```python
category_sales = df.groupBy("category").sum("revenue")

category_sales.show()
```

```python
avg_price = df.groupBy("product").avg("price")

avg_price.show()
```
